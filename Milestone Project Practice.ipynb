```python
def user_choice():
    choice = 'Default'
    
    while choice not in ['0','1','2']:
        
        choice = input('Enter a Index value 0 1 2 : ' )
        
        if choice not in ['0','1','2']:
            print('Sorry! , Please enter correct index values')
    
    return int(choice)
            
        
        
```

```python
def replacement_value(gamevalue):
    
    user_value = input('Enter the value that needs to changed :')
    
    position = user_choice()
    gamevalue[position] = user_value
    
    return gamevalue
```

```python
def game_play():
    choice = ''
    
    while choice not in ['Y','N']:
        
        choice = input('Do you wanna continue playing ? Y/N : ' )
        if choice not in ['Y','N']:
            print('Sorry! , Please Enter Y or N ')
    
    if choice == 'Y':
        return True
    else:
        return False
```
