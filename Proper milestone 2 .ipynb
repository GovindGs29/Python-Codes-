```python
class Card:
    
    def __init__(self,shape,rank):
        self.rank = rank
        self.shape = shape
        
    def __str__(self):
        return self.rank + " Of " + self.shape
```

```python
class Deck:
    
    def __init__(self):
        self.deck =[]
        for shape in shapes:
            for rank in ranks:
                self.deck.append(Card(shape,rank))
                
    def __str__(self):
        
        deck_contents = ''
        
        for card in self.deck:
            deck_contents +='\n' +card.__str__()
        return 'The Deck consists of :' + deck_contents
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        one_taken_card = self.deck.pop()
        return one_taken_card
                
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        
        if card.rank =="Ace":
            self.aces += 1
            
    def Ace_turn(self):
        while self.value > 21 and self.aces:
        # while we can use self.aces > 0
        # by default 0 means False and any number means True 
            self.value -=10
            self.aces -= 1
            
```
